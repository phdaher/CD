## Tabelas

```python
#Se usar o arquivo descompactado
tabela2015= pd.read_csv('DOM2015.csv',sep=',')

#Apenas as colunas que eu preciso no primeiro instante
cut=tabela2015.iloc[:,[1,2,28,47,49,55,57,78]]

#Ano da PNAD
cut=cut.rename(columns={'V0101': 'Ano'})

#Apenas os estados do Centro-Oeste
cut=cut[cut.UF>49]

#Coluna mostrando regão
cut.loc[cut.UF>49, "Região"]="Centro-Oeste"

#Nomeando Coluna Celulares
cut=cut.rename(columns={'V0220': 'Celular'})
cut.Celular=cut.Celular.astype("category")
cut.Celular.cat.categories=('Sim','Não')

#Nomeando coluna Internet nos Computadores
cut=cut.rename(columns={'V0232': 'NetComp'})
cut.NetComp=cut.NetComp.astype("category")
cut.NetComp.cat.categories=('Sim','Não')

#Nomeando coluna Internet em Casa
cut=cut.rename(columns={'V02322': 'NetCasa'})
cut.NetCasa=cut.NetCasa.astype("category")
cut.NetCasa.cat.categories=('Sim','Não')


cut=cut.rename(columns={'V02426': 'BandaLarga'})
cut.BandaLarga=cut.BandaLarga.astype("category")
cut.BandaLarga.cat.categories=('Sim','Não')

#Nomenado coluna Internet Movel
cut=cut.rename(columns={'V02424': 'NetMovel'})
cut.NetMovel=cut.NetMovel.astype("category")
cut.NetMovel.cat.categories=('Sim','Não')

#Nomeando coluna das Faixas de Salários Mínimos.
cut=cut.rename(columns={'V4622': 'FaixaRenda'})
cut.FaixaRenda=cut.FaixaRenda.astype("category")
cut.FaixaRenda.cat.categories=('SemRend','1/4 ','1/4 até 1/2','1/2 até 1','1 até 2','2 até 3','3 até 5', 'Mais de 5', 'NãoDec')

#Nomeando os estados que vou utilizar
cut=cut.rename(columns={'UF': 'Estado'})
cut.Estado=cut.Estado.astype('category')
cut.Estado.cat.categories=('Mato Grosso do Sul', 'Mato Grosso', 'Goiás',  'Distrito Federal')

#imprimindo apenas os 5 primeiros valores da tabela
cut.head()


#Os Nans não foram tirados do meu trabalho pois haviam vários Nans e se eu os removesse comprometeria os meus resultados, analises e gráficos
#cut=cut.dropna()


```
